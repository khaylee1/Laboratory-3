```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

##<b>Task 1
Try different linear combinations using different scalar values.<br> 

$$K= \begin{Bmatrix}a\cdot\begin{bmatrix} -4\\3 \\\end{bmatrix}, 
b\cdot \begin{bmatrix} 7\\2 \\\end{bmatrix},c\cdot\begin{bmatrix}-5\\4\\\end{bmatrix}\end{Bmatrix} $$

Code and plot:

```
x=np.array([-4,3])
y=np.array([2,2])
z=np.array([-4,-1])

a=np.arange(-10,10,0.25)
b= np.arange(-10,10,2)
c=np.arange(-10,10,0.5)

plt.scatter(a*x[0],a*x[1],label='span of vector A',color='orange',)
plt.scatter(b*y[0],b*y[1],label='span of vector B',color='yellow')
plt.scatter(c*z[0],c*z[1],label='span of vector C',color='blue')

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.legend()
plt.show()
```

## <B>Task 2
Try making three unque spans using different linear combinations 
$$S = \begin{Bmatrix} c_1 \cdot\begin{bmatrix} -3\\-3 \\\end{bmatrix}, 
c_2 \cdot \begin{bmatrix} 2\\2 \\\end{bmatrix},c_3\cdot\begin{bmatrix}3\\-4\\\end{bmatrix}\end{Bmatrix} $$
$$K= \begin{Bmatrix} c_1 \cdot\begin{bmatrix} 1\\2 \\\end{bmatrix}, 
c_2 \cdot \begin{bmatrix} 3\\4 \\\end{bmatrix},c_3\cdot\begin{bmatrix}5\\-6\\\end{bmatrix}\end{Bmatrix} $$
$$J= \begin{Bmatrix} c_1 \cdot\begin{bmatrix} 1\\5 \\\end{bmatrix}, 
c_2 \cdot \begin{bmatrix} -4\\2 \\\end{bmatrix},c_3\cdot\begin{bmatrix}3\\-6\\\end{bmatrix}\end{Bmatrix} $$

<B>code and plot:

```
fig_1=plt.figure(1,figsize=(20,9.6))
chart_1=fig_1.add_subplot(221)
#vectors
vectx = np.array([-3,-2])
vecty = np.array([2,2])
vectz= np.array([3,-4])
R = np.arange(-10,10,1)
c1, c2, c3 = np.meshgrid(R,R,R)


spanx = c1*vectx[0] + c2*vecty[0]+c3*vectz[0]
spany = c1*vectx[1] + c2*vecty[1]+c3*vectz[1]

#span of the 3 given vectors
plt.scatter(R*vectx[0],R*vectx[1],color='y')
plt.scatter(R*vecty[0],R*vecty[1], color='y')
plt.scatter(R*vectz[0],R*vectz[1],color='y') 

#appearance of the plot 
plt.scatter(spanx,spany, s=5, alpha=0.75,color='b')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()

chart_2=fig_1.add_subplot(222)
#vectors
vecta = np.array([1,2])
vectb = np.array([3,4])
vectc= np.array([5,6])
R = np.arange(-10,10,1)
c1, c2, c3 = np.meshgrid(R,R,R)


spanx = c1*vecta[0] + c2*vectb[0]+c3*vectc[0]
spany = c1*vecta[1] + c2*vectb[1]+c3*vectc[1]

#span of the 3 given vectors
plt.scatter(R*vecta[0],R*vectb[1],color='r')
plt.scatter(R*vectb[0],R*vectb[1],color='r')
plt.scatter(R*vectc[0],R*vectc[1],color='r')

#appearance of the plot 
plt.scatter(spanx,spany, s=5, alpha=0.75,color='y')


#code for the chart 
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()

chart_3=fig_1.add_subplot(223)
#vectors
vecta1 = np.array([1,5])
vectb1 = np.array([-4,2])
vectc1= np.array([3,-6])
R = np.arange(-10,10,1)
c1, c2, c3 = np.meshgrid(R,R,R)


spanx = c1*vecta1[0] + c2*vectb1[0]+c3*vectc1[0]
spany = c1*vecta1[1] + c2*vectb1[1]+c3*vectc1[1]

#span of the 3 given vectors
plt.scatter(R*vecta1[0],R*vectb1[1],color='r')
plt.scatter(R*vectb1[0],R*vectb1[1],color='r')
plt.scatter(R*vectc1[0],R*vectc1[1],color='r')

#appearance of the plot 
plt.scatter(spanx,spany, s=5, alpha=0.75,color='orange')


#code for the chart 
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()

#code for the chart 
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()

plt.show()
```
