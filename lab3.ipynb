##<b>Task 1
Try different linear combinations using different scalar values.<br> 

$$K= \begin{Bmatrix}a\cdot\begin{bmatrix} -4\\3 \\\end{bmatrix}, 
b\cdot \begin{bmatrix} 7\\2 \\\end{bmatrix},c\cdot\begin{bmatrix}-5\\4\\\end{bmatrix}\end{Bmatrix} $$

Code and plot:

```
x=np.array([-4,3])
y=np.array([2,2])
z=np.array([-4,-1])

a=np.arange(-10,10,0.25)
b= np.arange(-10,10,2)
c=np.arange(-10,10,0.5)

plt.scatter(a*x[0],a*x[1],label='span of vector A',color='orange',)
plt.scatter(b*y[0],b*y[1],label='span of vector B',color='yellow')
plt.scatter(c*z[0],c*z[1],label='span of vector C',color='blue')

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.legend()
plt.show()
```
